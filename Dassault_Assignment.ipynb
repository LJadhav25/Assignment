#**a.** Create a plot to find out if there is any co-relation between the Blood Group Type, Gender and the Medical Condition?

```python
data_encoded = pd.get_dummies(health_care[['Gender', 'Blood Group Type', 'Medical Condition']])
correlation_matrix = data_encoded.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.10,vmin=-1, vmax=1)
```

#b. Find out the Average Billing Amount required to treat each medical Condition?

```python
health_care.groupby('Medical Condition')['Billing Amount'].mean()
```
